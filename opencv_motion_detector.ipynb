```python
import cv2,time,pandas
from datetime import datetime
static_back=None
# global variable to define for the rest of function
currentmotion=["",""]
#to store the current and futute time and calculate the motion on basis of it
motionList=[]
# to store the 
video=cv2.VideoCapture(0)
#to get the camera
while True:
    check,frame=video.read()
    #for reading each and every frame
    motion=0
    #initialisng motion to zero
    #converting color to gray color
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    #converting gaussinn blur
    gray=cv2.GaussianBlur(gray,(21,21),0)
    if static_back is None:
        static_back=gray
        continue
    #getting differnce between the gaussian frame and current frame
    diff_frame=cv2.absdiff(static_back,gray)
    #if the abs diffenrce between static frame and dynamic is 30
    #it will show white color
    thresh_frame=cv2.threshold(diff_frame,30,255,cv2.THRESH_BINARY)[1]
    thresh_frame=cv2.dilate(thresh_frame,None)
    #finding contours in the next frame
    cnts,_=cv2.findContours(thresh_frame.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    for contour in cnts:
        if cv2.contourArea(contour)<10000:
            motion=1
            continue
        if currentmotion[0]=="" and motion==1:
            currentmotion[0]=datetime.now().strftime("%m/%d/%y,%H:%M:%S")
        # creating a rectangle around the bound contours
        (x,y,w,h)=cv2.boundingRect(contour)
        #making green rectangle around the moving object
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
        #m appending status of motion
    if motion==0 and currentmotion[0]!="":
        currentmotion[1]=datetime.now().strftime("%m/%d/%y,%H:%M:%S")
        motionlist.append(currentmotion)
        currentmotion=["",""]
    #display gray image
    cv2.imshow("gray Imae",gray)
    # display the diffential frame
    cv2.imshow("differential frame",diff_frame)
    
        
        
        
        
            
    
    
    key=cv2.waitKey(1)
    
    # if q enters whole process will stop
    #if the first iteration if value of static_back is none initailizw it to frame
    # if something is moving then append the end of motionlist
    
    if key==ord('q'):
        if motion==1 and  currentmotion[0]!="":
            currentmotion[1]=datetime.now().strftime("%m/%d/%y,%H:%M:%S")
            motionList.append(currentmotion)
        break
for move in motionList[:30]:
    print(move)
video.release()
cv2.distroyAllWindows()
    
    
```
